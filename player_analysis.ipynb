```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from datasets import matches_by_players
matches_by_players_df = matches_by_players()
```

# Análisis preliminar players

```python
# calcular el promedio de la distancia total de movimiento por ganador
promedio_distancia = matches_by_players_df.groupby('winner')['movement_total_distance'].mean()
# Gráfico
plt.bar(['Perdidos', 'Ganados'], promedio_distancia)
# Etiquetas
plt.xlabel('Perdidos/Ganados')
plt.ylabel('Distancia promedio de movimiento')

plt.show()
```

```python
# En las siguientes líneas de código se crea un gráfico de histograma para mostrar la relación entre la distancia recorrida y los partidos ganados o perdidos.
fig,ax = plt.subplots(figsize=(6,4))
# Se crea un histograma para los partidos perdidos con color rojo y ganados con color verde
matches_by_players_df[matches_by_players_df['winner'] == False]['movement_total_distance'].hist(alpha=0.5, label='Partidos perdidos', color='red')
matches_by_players_df[matches_by_players_df['winner'] == True]['movement_total_distance'].hist(alpha=0.5, label='Partidos ganados', color='green')
ax.set_title(label='Histograma relación distancia recorrida y partidos ganados o perdidos', fontsize=14, x=0.5)
plt.legend()
plt.show()
```

```python
# En las siguientes líneas de código se crea un gráfico de histograma para mostrar la relación entre el boost recolectado y los partidos ganados o perdidos.
fig,ax = plt.subplots(figsize=(6,4))
# Se crea un histograma para los partidos perdidos con color rojo y ganados con color verde
matches_by_players_df[matches_by_players_df['winner'] == False]['boost_amount_collected'].hist(alpha=0.5, label='Partidos perdidos', color='red')
matches_by_players_df[matches_by_players_df['winner'] == True]['boost_amount_collected'].hist(alpha=0.5, label='Partidos ganados', color='green')
ax.set_title(label='Histograma relación boost recolectado y partidos ganados o perdidos', fontsize=14, x=0.5)
plt.legend()
plt.show()
```
